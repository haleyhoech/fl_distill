```python
import torch, torchvision
import numpy as np
import data
from data import get_cifar10
from data import split_image_data
```

```python
train_data, test_data = get_cifar10("~/Data/PyTorch/")
```

```python
labels = train_data.targets
n_clients = 10
classes_per_client = 2
batch_size =128
n_data = None
```

```python
n_data = len(labels)
n_labels = np.max(labels) + 1
```

```python
label_idcs = {l : np.argwhere(np.array(labels[:n_data])==l).flatten().tolist() for l in range(n_labels)}
label_idcs
```

```python
data_per_client = n_data // n_clients
data_per_client
```

```python
data_per_client_per_class = data_per_client // classes_per_client
data_per_client_per_class
```

```python
idcs = []
for i in range(n_clients):
    client_idcs = []
    budget = data_per_client
    c = np.random.randint(n_labels)
    while budget > 0:
        take = min(data_per_client_per_class, len(label_idcs[c]), budget)
        
        client_idcs += label_idcs[c][:take]
        label_idcs[c] = label_idcs[c][take:]
        
        budget -= take
        c = (c + 1) % n_labels

    idcs += [client_idcs]
```
